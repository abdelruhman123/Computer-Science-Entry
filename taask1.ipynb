Handling Missing Values

```python
df = df.dropna(subset=['lat', 'long', 'price'])
```

```python
df.isnull().any()
```

```python
# i will delete (country and country code ) because all of then is united states
df = df.drop (columns=['country', 'country code'])
```

```python
df = df.dropna(subset=['lat', 'long', 'price'])
```

```python
df ['service fee'] = df ['service fee'].fillna(0)
```

```python
# loobing on the textst columns forn filling the missing value

cols = ['host_identity_verified', 'host name', 'NAME', 'house_rules','neighbourhood group','neighbourhood']
for col in cols:
    df [col] = df [col]. fillna('unknow').astype('string')
```

```python
df ['house_rules'] = df['house_rules']. replace ('#NAME?', 'unknown')
```

```python
# remove the $ from price  before loobing
df['price'] = df['price'].str.replace('$', '').str.replace(',', '').astype(float)
df['service fee'] = df['service fee'].str.replace('$', '').str.replace(',', '').astype(float, errors='ignore')
## loobing on numric value
numeric_columns = {
    'price': 0,
    'minimum nights': 1,
    'number of reviews': 0,
    'reviews per month': 0,
    'review rate number': df['review rate number'].median(),
    'calculated host listings count': 1,
    'availability 365': 0,
    'Construction year': df['Construction year'].median()
}

for col, default_value in numeric_columns.items():
    df[col] = df[col].fillna(default_value).astype('int32' if col != 'review rate number' else 'float64')
```
